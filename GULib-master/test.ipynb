```python
import networkx as nx
import numpy as np

def test_adj_matrix():
    # 创建一个简单的图
    G = nx.karate_club_graph()  # 使用 Karate Club 图作为示例
    print("原始图的节点数:", G.number_of_nodes())
    print("原始图的边数:", G.number_of_edges())

    # 调用 adj_matrix 函数
    # 在较新版本的 networkx 中，adj_matrix 函数位于 convert_matrix 模块中
    adj_matrix = nx.adjacency_matrix(G).toarray().astype(np.bool_)

    # 打印邻接矩阵
    print("邻接矩阵:")
    print(adj_matrix)

    # 验证邻接矩阵的形状
    assert adj_matrix.shape == (G.number_of_nodes(), G.number_of_nodes()), "邻接矩阵的形状不正确"

    print("测试通过")
```

```python
test_adj_matrix()
```
